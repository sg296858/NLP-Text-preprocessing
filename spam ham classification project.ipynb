```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['Category']=label.fit_transform(df['Category'])
```

```python
df['Message']=df['Message'].str.lower()
```

```python
df['Message']
```

```python
def remove_htmltags(text):
    import re
    pattern=re.compile('<.*?>')
    return pattern.sub(r'',text)
```

```python
df['Message']=df['Message'].apply(remove_htmltags)
```

```python
def remove_urls(text):
    import re
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)
```

```python
df['Message']=df['Message'].apply(remove_urls)
```

```python
def remove_punct(text):
    import string
    for punctuation in string.punctuation:
        text=text.replace(punctuation,'')
    return text
```

```python
df['Message']=df['Message'].apply(remove_punct)
```

```python
import emoji
def replace_emojis(text):
  return emoji.demojize(text)
```

```python
df['Message']=df['Message'].apply(replace_emojis)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
nltk.download('punkt_tab')
```

```python
def tokenize(text):
  return word_tokenize(text)
```

```python
df['message_tokenize']=df['Message'].apply(tokenize)
```

```python
df
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def remove_stopwords(text):
  lemmas=[lemmatizer.lemmatize(word) for word in text if word not in stopwords.words('english')]
  return " ".join(lemmas)
```

```python
df['message_tokenize']=df['message_tokenize'].apply(remove_stopwords)
```

```python
df
```

```python
X=df['message_tokenize']
Y=df[['Category']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from imblearn.over_sampling import RandomOverSampler
rs=RandomOverSampler()
X_train_rs,Y_train_rs=rs.fit_resample(np.array(X_train).reshape(-1,1),Y_train)
```

```python
X_train_rs.shape
```

```python
X_train.shape
```

```python
Y_train_rs.shape
```

```python
Y_train.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
X_train_tfidf=tfidf.fit_transform(X_train_rs.ravel())
X_test_tfidf=tfidf.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train_tfidf,Y_train_rs)
```

```python
y_pred=model.predict(X_test_tfidf)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(Y_test,y_pred)
```

```python
classification_report(Y_test,y_pred)
```

```python
from sklearn.naive_bayes import MultinomialNB
model1=MultinomialNB()
model1.fit(X_train_tfidf,Y_train_rs)
```

```python
model1.score(X_train_tfidf,Y_train_rs)
```
