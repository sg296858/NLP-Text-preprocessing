```python
df['review']=df['review'].str.lower()
```

```python
def remove_htmltags(text):
    import re
    pattern=re.compile('<.*?>')
    return pattern.sub(r'',text)
```

```python
df['review']=df['review'].apply(remove_htmltags)
```

```python
df['review']
```

```python
def remove_urls(text):
    import re
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)
```

```python
df['review']=df['review'].apply(remove_urls)
```

```python
def remove_punctuations(text):
    import string
    for punctuation in string.punctuation:
        text=text.replace(punctuation,'')
    return text
```

```python
df['review']=df['review'].apply(remove_punctuations)
```

```python
import emoji
def replace_emojis(text):
  return emoji.demojize(text)
```

```python
df['review']=df['review'].apply(replace_emojis)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
def tokenize(text):
  return word_tokenize(text)
```

```python
df['review_tokenize']=df['review'].apply(tokenize)
```

```python
df
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def remove_stopwords(text):
  lemmas=[lemmatizer.lemmatize(word) for word in text.split() if word not in stopwords.words('english')]
  return " ".join(lemmas)
```

```python
df['review_removedstopwords']=df['review'].apply(remove_stopwords)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['sentiment']=label.fit_transform(df['sentiment'])
```

```python
df
```

```python
X=df['review_removedstopwords']
Y=df[['sentiment']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
X_train_cv=cv.fit_transform(X_train)
X_test_cv=cv.transform(X_test)
```

```python
X_train_cv
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_cv,Y_train)
```

```python
y_pred=model.predict(X_test_cv)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(X_train_cv,Y_train)
```

```python
y_pred=model1.predict(X_test_cv)
y_pred
```

```python
accuracy_score(Y_test,y_pred)
```
