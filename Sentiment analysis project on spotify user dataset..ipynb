```python
from nltk.tokenize import word_tokenize

def tokenize(text):
    return word_tokenize(text)
    


df['Review'] = df['Review'].apply(tokenize)
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

def remove_stopwords(text):
    stemmer = PorterStemmer()
    stem_words = [stemmer.stem(word) for word in text if word not in stopwords.words('english')]
    return " ".join(stem_words)
    
df['Review'] = df['Review'].apply(remove_stopwords)
```

```python
df['Review'][100]
```

```python
import emoji

def replace_emojis(text):
    return emoji.demojize(text)
    

# Apply the function to the 'Review' column
df['Review'] = df['Review'].apply(replace_emojis)
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['label']=label.fit_transform(df['label'])
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def tfidf_vectorize(text):
    vectorizer = TfidfVectorizer()
    vector = vectorizer.fit_transform(text)
    return vector
```

```python
df['Review'] = df['Review'].astype(str)
```

```python
mat=tfidf_vectorize(df['Review'])
```

```python
mat.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(mat,df['label'],random_state=1,test_size=0.2)
```

```python
X_train
```

```python
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_pred,Y_test)
```
