```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({'text':['people watch campusx','campusx watch campusx','people write comments','campusx write comment'],'output':[1,1,0,0]})
```

```python
df.head()
```

#BOW implementation n_grams,bigrams,trigrams

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(ngram_range=(2,2))
bow=cv.fit_transform(df['text']).toarray()
```

```python
bow
```

```python
cv.vocabulary_
```

#TF-IDF implementation

```python
df1=pd.DataFrame({'text':['people watch campusx','campusx watch campusx','people write comments','campusx write comment'],'output':[1,1,0,0]})
```

```python
df1.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
bow1=tfidf.fit_transform(df1['text']).toarray()
```

```python
bow1
```

```python
tfidf.idf_
```

```python
tfidf.get_feature_names_out()
```
